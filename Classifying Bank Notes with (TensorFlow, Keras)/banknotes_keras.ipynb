## DATA PREPARATION
---

```python
X = df.drop('Class', axis=1)
y = df['Class']

#scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X)
```

```python
#Converting the scaled_features back to original df
df_scaled_features = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

## PREDICTIVE MODELLING
---

#### SPLIT THE DATA

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### KERAS MODEL
---

```python
#building the keras Sequential model
model = Sequential()
model.add(Dense(units = 100 , activation = 'relu', input_dim=X_train.shape[1]))
model.add(Dense(units = 50 , activation = 'relu'))
model.add(Dense(units = 25 , activation = 'relu'))
model.add(Dense(units = 10 , activation = 'relu'))
model.add(Dense(units = 1 , activation = 'sigmoid'))
model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
model.summary()
```

```python
#train the model with 5 epochs
model.fit(X_train, y_train, epochs=5)
```

```python
#Get predictions of X_test
predictions = model.predict_classes(X_test)
predictions = predictions.reshape(predictions.shape[0])
```

```python
#print Classification report
print(classification_report(y_test, predictions))
```

* #### The model looks good, it was able to predict the classes perfectly with 100% accuracy!

### XGBOOST CLASSIFIER
---

```python
from scipy.stats import uniform

#RandomSearchCV
# Hyperparameters to tune and their ranges
param_dist = {"learning_rate": uniform(0, 2),
              "gamma": uniform(1, 0.000001),
              "max_depth": range(1,50),
              "n_estimators": range(1,300),
              "min_child_weight": range(1,10),
              'n_jobs': range(1,5)}
```

```python
#instance of RansearchCV
rs = RandomizedSearchCV(XGBClassifier(), param_distributions=param_dist, n_iter=25) #25 iterations
#fit the model
rs.fit(X_train, y_train)
```

```python
#Predict the test data
predictions = rs.predict(X_test)
#accuracy test
xgb_accu = accuracy_score(y_test, predictions)
print('Model Accuracy: ', round(xgb_accu*100, 2), '%')
```

```python
#print classification report
print(classification_report(y_test, predictions))
```

### INSIGHT
---
* XGBClassifier model perfomed well, though keras model is a bit better here when classifying bank notes if they are either fake or real.
