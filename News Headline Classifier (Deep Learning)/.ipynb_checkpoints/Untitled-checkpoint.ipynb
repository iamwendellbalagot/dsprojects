## TEXT PREPROCESSING
---

```python
#remove the empty word headlines
df = df[df['lenght_text']>=5]
```

#### TEXT TOKENIZING

```python
#tokenizing the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df.text)
#make it as our training feature
X = tokenizer.texts_to_sequences(df.text)
df['words'] = X
```

#### TEXT TO SEQUENCES

```python
#get maxlen on X array
maxlen = -1
for i in X:
    if maxlen<len(i):
        maxlen = len(i)
```

```python
#padding
X_padded = sequence.pad_sequences(X,maxlen=maxlen, padding='post') 
```

#### LABEL TRANSFORM THE CATEGORIES

```python
#rename The worldpost to worldpost since they are the same thing
df.category = df.category.map(lambda x: "WORLDPOST" if x == "THE WORLDPOST" else x)
#get the labels
categories = df.groupby('category').size().index
```

```python
#label encode the categories on dataframe
le = LabelEncoder()
labels = le.fit_transform(df.category)
df['labels'] = labels
#map the categories
leb = LabelEncoder()
cat = le.fit_transform(categories)
```

#### LOAD AND PREPROCESS THE GLOVE EMBEDDING

```python
#get the word vocabolary index
word_index = tokenizer.word_index
#Embedding vector dimension
EMBEDDING_DIM = 100

embedding_index = {}
file = open('data/glove.6B.100d.txt', encoding='utf8')

for line in file:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embedding_index[word] = coefs
file.close()
```

```python
#create embedding matrix
embedding_matrix = np.zeros(shape=(len(word_index)+1,EMBEDDING_DIM))
```

```python
#search the word vocabulary items on the embeddings
for word, index in word_index.items():
    embedding_vector = embedding_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[index] = embedding_vector
```

#### CREATE THE EMBEDDING LAYER

```python
#load the pre-trained embedding layer
embedding_layer = Embedding((len(word_index)+1), EMBEDDING_DIM,
                            embeddings_initializer=Constant(embedding_matrix),
                            input_length=maxlen,
                            trainable=False)
```

## PREDICTIVE MODELLING
---

```python
#converting y to keras categorical values
y = to_categorical(labels)
#splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=101)
```

### KERAS SEQUENTIAL MODEL

```python
model = Sequential()
model.add(embedding_layer)
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(cat), activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#print summary
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=5, batch_size=128)
```

#### LOSS AND ACCURACY

```python
#plot the loss and accuracy
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(history.history['loss'], color='red')
ax[1].plot(history.history['accuracy'], color='green')
ax[0].set_title('TRAINING LOSS', fontsize=16)
ax[1].set_title('TRAINING ACCURACY', fontsize=16)
```

#### PREDICT THE TEST DATA

```python
#get predictions
predictions = model.predict_classes(X_test)
```

#### CLASSIFICATION REPORT

```python
#classification report
print(classification_report(y_test.argmax(1), predictions))
```

```python
#print test accuracy
print('Model Accuracy: ', round(accuracy_score(y_test.argmax(1), predictions)*100,2), '%')
```
