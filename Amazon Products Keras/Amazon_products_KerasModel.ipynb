## DATA PREPARATION
---

```python
#label encode the variation
df['variation'] = LabelEncoder().fit_transform(df.variation)
```

```python
#drop date and feedback feature
df.drop(['date', 'feedback'], axis=1, inplace=True)
```

```python
#Scale the text length feature
scaler = StandardScaler()
df['text_lenght'] = scaler.fit_transform(df.text_lenght.values.reshape(-1,1))
```

```python
X = df.drop('rating', axis=1)
y = df['rating']

#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## PREDICTIVE MODELLING
---

### KERAS MODEL
---

```python
#specify the number of vocabolary
max_words = 500
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train.verified_reviews)
train_tokenize = tokenizer.texts_to_sequences(X_train.verified_reviews)
```

```python
#get the max lenght of every reviews
maxlen = -1

for text in train_tokenize:
    if (maxlen<len(text)):
        maxlen = len(text)
```

```python
#padding
train_padded = pad_sequences(train_tokenize, maxlen=maxlen, padding='post')
#padding test data
test_tokenize = tokenizer.texts_to_sequences(X_test.verified_reviews)
test_padded = pad_sequences(test_tokenize, maxlen=maxlen, padding='post')
```

```python
#convert labels to categorical value
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
#Building the sequential model
model = Sequential()
model.add(Embedding(max_words,64,input_length = maxlen))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(6, activation='softmax'))


model.compile(optimizer ='adam' , loss = 'categorical_crossentropy' , metrics = ['accuracy'])
model.summary()
```

```python
# TRAIN THE DATA WITH 5 EPOCHs
history = model.fit(train_padded,y_train,epochs=20, batch_size=32)
```

```python
#plot the model training loss
plt.plot(history.history['loss'], label='train')
plt.title('MODEL LOSS AROUND 20 EPOCHS')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
#Evaluate the test data and get the accuracy
lstm_accu = model.evaluate(test_padded, y_test)[1]
```

```python
print('Model Accuracy: ', round(lstm_accu*100, 2), '%')
```

* #### Model Accuracy can be further improve by increasing the training epochs.
